```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Available dataset in seaborn
```

```python
sns.get_dataset_names()
```

```python
tips =sns.load_dataset("tips")
tips
```

# To Calculate tips for a day 

```python
tips.groupby('day').mean(numeric_only = True)
```

```python
sns.barplot(tips ,x='day',y='tip',  hue ='sex' )
# this is reduced to mean  we can change using estimator default is mean  we can do like , estimator = 'median'
```

```python
sns.barplot(tips ,x='sex',y='tip',errorbar =('ci', 95) , hue ='day'  )
```

```python

sns.barplot(tips ,x='total_bill',y='day',errorbar =('ci', 95) , hue ='sex', palette ='viridis', orient ='h',estimator ="median")
```

```python
sns.histplot(tips, x ="total_bill",kde =True)
# kde = kernel density estimate i.e bar length are approximated as we see from line and not through histogram so it kind of generalise the plot
```

```python
sns.histplot(tips, x ="total_bill", y ="tip",kde =True) 
# this accumulates hte data set to depict where the most value lies i.e it is the density of the distribution i.e darker region is more data Pontas
```

```python
sns.kdeplot(tips, x = "tip")
```

```python
sns.kdeplot(tips, x = "tip", y = "total_bill")
# counter depicts like for  eg it is hillsand more dense area depicts more  circles
```

```python
fmri_data =sns.load_dataset('fmri')
fmri_data
```

```python
sns.lineplot(fmri_data, x = "timepoint", y ="signal")
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(fmri_data, x = "timepoint", y ="signal", hue ="event", style="region", markers=["*",'o'])
```

```python
sns.jointplot(tips, x = "total_bill", y = "tip", hue ="sex")
```

```python
sns.jointplot(tips, x = "total_bill", y = "tip", kind = "reg")
```

```python
sns.pairplot(tips, hue ='day')
```

```python
sns.boxplot(tips, x = 'day' ,y ='total_bill', hue ='sex')
```

```python
sns.violinplot(tips,x ='day', y = 'total_bill')
# doesn't tell abt outliers but tells about distribution
```

```python
sns.violinplot(tips,x ='day', y = 'total_bill', split= True, hue = 'sex')
```

```python
sns.swarmplot(tips, x = 'day', y = 'tip')
```

```python
sns.stripplot(tips, x = 'day', y = 'tip')
```
