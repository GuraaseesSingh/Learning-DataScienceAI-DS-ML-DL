### Population vs year

```python
import pandas as pd
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('populations.txt', sep ='\t')
# print(df)
carrot = df[ 'carrot']
year = df['# year']
print(df.columns)
print(carrot)
print(year)
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(year,carrot)
plt.grid()
plt.show()
```
